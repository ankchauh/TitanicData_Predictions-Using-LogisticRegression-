```python
train_data = pd.read_csv("titanic_x_y_train.csv", skipinitialspace=True)
test_data = pd.read_csv("titanic_x_test.csv", skipinitialspace=True)
```

```python
train_data.drop(columns=['Name','SibSp','Parch','Ticket','Fare','Cabin'], inplace=True)
```

```python
train_data['Age'][train_data['Age'].isnull()]=train_data['Age'].mean()
train_data['Embarked'][train_data['Embarked'].isnull()]='S'
```

```python
train_data.isnull().sum()
```

```python
train_data['Sex'][train_data['Sex']=='male']=1
train_data['Sex'][train_data['Sex']=='female']=0
train_data['Embarked'][train_data['Embarked']=='S']=1
train_data['Embarked'][train_data['Embarked']=='C']=2
train_data['Embarked'][train_data['Embarked']=='Q']=3
```

```python
train_data.head(50)
```

```python
np_train = train_data.to_numpy(dtype=int, copy=False)
```

```python
np_train.shape
```

```python
x_train=np_train[:,:-1]
y_train=np_train[:,-1]
```

```python
print(x_train.shape, y_train.shape)
```

```python
scaler=preprocessing.StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
```

```python
x_train
```

```python
alg=LogisticRegression()
alg.fit(x_train, y_train)
```

```python
test_data.head()
```

```python
test_data.isnull().sum()
```

```python
test_data.drop(columns=['Name','SibSp','Parch','Ticket','Fare','Cabin'], inplace=True)
test_data['Age'][test_data['Age'].isnull()]=test_data['Age'].mean()
test_data['Embarked'][test_data['Embarked'].isnull()]='S'
test_data['Sex'][test_data['Sex']=='male']=1
test_data['Sex'][test_data['Sex']=='female']=0
test_data['Embarked'][test_data['Embarked']=='S']=1
test_data['Embarked'][test_data['Embarked']=='C']=2
test_data['Embarked'][test_data['Embarked']=='Q']=3
np_test = test_data.to_numpy(dtype=int, copy=False)
```

```python
np_test.shape
```

```python
x_test=scaler.transform(np_test)
```

```python
result = alg.predict(x_test)
```
